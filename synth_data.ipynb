# Synthetic Data Generator

This notebook generates the synthetic data for the LSTM-AE. These results become the time-series (both normal and anomaly) used in that notebook. Try playing with the values and functions here before rerunning the LSTM-AE!

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

A few constants we want to keep in a convienent place

```python
# Useful things
noise_level = 0.2
```

## Normal Time-Series

```python
# Nominal case (stochastic)
df = pd.DataFrame()
dfd = pd.DataFrame()

x1_1 = np.linspace(0, 5, num=50)
x1_2 = np.linspace(5, 15, num=50)
x1_3 = np.linspace(15, 20, num=50)
x1 = np.concatenate((x1_1,x1_2,x1_3))
y_normal_1 = 5*x1_1/np.max(x1_1)
y_normal_2 = 0.25*np.sin(x1_2) + 5
y_normal_3 = -5*x1_1/np.max(x1_1) + 5
y_normal = np.concatenate((y_normal_1,y_normal_2,y_normal_3))

y = abs(y_normal)
p =  np.sqrt(np.squeeze(y))

for i in range(1000):
    w = np.random.normal(scale=noise_level, size=linspace_length)
    y2 = y_normal + w
    cat = "normal"

    df_a = pd.DataFrame([cat])
    df = pd.concat([df, df_a])
    ser = pd.Series(y2)
    dfc = ser.to_frame().T
    dfd = pd.concat([dfd,dfc])

df = pd.concat([df, dfd],axis=1)
df.to_csv('synthetic_data_normal.csv', sep=',',index=False)

# Plotting
sns.set_theme()
plt.figure()
plt.plot(x1,y_normal)
plt.plot(x1,y_normal + p,'k--')
plt.plot(x1,y_normal-p,'k--')
plt.legend(["normal","+/-1 std",])
plt.title("Deterministic Synthetic Data Model")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")

plt.figure()
for j in range(100):
    plt.plot(x1, df.iloc[j, 1:])
plt.plot(x1,y_normal + p,'k--')
plt.plot(x1,y_normal-p,'k--')
#plt.legend(["normal","high","low","saturation","off"])
plt.title("Stochastic Synthetic Data Model")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")

plt.show()
```
